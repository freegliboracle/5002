```python
import numpy as np
import pandas as pd
import datetime
import os
from dateutil.parser import parse
from datetime import timedelta, datetime
```

```python
cache_path = './cache/'
data_path = './data/'
grid_stations = pd.read_csv(data_path + 'Beijing_grid_weather_station.csv')
```

```python
grid_stations.head()
```

```python
def date_add_hours(start_date, hours):
    end_date = parse(start_date) + timedelta(hours=hours)
    end_date = end_date.strftime('%Y-%m-%d %H:%M:%S')
    return end_date
```

```python
meo_grid = pd.read_hdf(data_path + 'meo_grid.hdf')
```

```python
meo_grid.head()
```

```python
min_time = meo_grid['time'].min()
max_time = meo_grid['time'].max()
min_time, max_time
```

```python
# dates = [str(date)[:10] for date in pd.date_range('2017-01-01','2017-01-06')]
dates = [str(date)[:10] for date in pd.date_range(min_time,max_time)]
for date in dates:
    start_time = date_add_hours(date, 0)[:10] #+ ' 23:00:00'
#     next 48 hours
    end_time = date_add_hours(date, 48)[:10] #+ ' 23:00:00'
    result_path = cache_path + "meo_grid_{}.csv".format(date)
    if not os.path.exists(result_path):
        try:
            result = meo_grid[(meo_grid['time']>=start_time) & (meo_grid['time']<end_time)].copy()
            result = grid_stations.merge(result, on='station_id', how='inner')
            result.to_csv(result_path, index=False)
            print('updating {}...'.format(date))
        except:
            print('updating {} fail'.format(date))
    else:
        print('{} data exists'.format(date))
```
