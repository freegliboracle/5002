```python
for csv in csv_list:
    result_path = cache_path+'aq_'+csv
    if not os.path.exists(result_path):
        weather = pd.read_csv(grid_meo_path+csv)
        result = []
        for key in aq_station_nearest_meo_station:
            result_temp=weather.replace(aq_station_nearest_meo_station[key], key)[weather['station_id']==aq_station_nearest_meo_station[key]]
            result.append(result_temp)
        feat = pd.concat(result,axis=0)
        feat.to_csv(result_path, index=False)
        print('{} updating'.format('aq_'+csv))
    else:
        print('{} exists'.format('aq_'+csv))
```
